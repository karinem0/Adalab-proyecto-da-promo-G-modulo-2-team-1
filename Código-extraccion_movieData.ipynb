```python
import requests
import pandas as pd 

# URL de la API y encabezados necesarios para realizar la solicitud
url = "https://moviesdatabase.p.rapidapi.com/titles"
headers = {
    "X-RapidAPI-Key": "b38cd0a830msh149cd093d780ffcp1fa141jsnc85ccd4fb846",
    "X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"
}

# Función para llamar a la API y obtener los datos de películas
def llamar_api(genero, tipo, year, page):
    querystring = {
        "genre": genero,
        "startYear": year,
        "endYear" : year,
        "titleType": tipo,
        "page": page
    }
    # Realizamos la solicitud GET a la API con los parámetros proporcionados
    response = requests.get(url, headers=headers, params=querystring)
    # Verificamos si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Solicitamos que nos devuelvan los resultados de la API en formato JSON
        return response.json()['results']
    else:
        # Nos mostrará en pantalla un mensaje de error si la solicitud no fue exitosa
        print(f"Error al obtener {tipo} de {genero} para el año {year}, página {page}: {response.text}")
        # Devolvemos una lista vacía en caso de error
        return []

# Función para obtener los datos de películas de varios años y páginas
def obtener_datos(genero, tipo, fecha_inicio, fecha_fin):
    lista_peliculas = []
    # Iteramos sobre los años que querramos obtener 
    for year in range(fecha_inicio, fecha_fin + 1):
        print(f"Obteniendo películas {tipo} de {genero} para el año {year}:")
        # Iterar sobre las páginas de resultados (máximo 50 páginas)
        for page in range(1, 51):
            # Llamamos a la función llamar_api para obtener datos de películas
            peliculas = llamar_api(genero, tipo, year, page)
            # Si no hay más películas en la página actual, salir del bucle interno
            if not peliculas:
                break
            # Iteramos para lograr película según nuestros parámetros y así gregarla a la lista de películas
            for pelicula in peliculas:
                if pelicula['titleType']['text'] == tipo.capitalize():
                    tipo_pelicula = pelicula['titleType']['text']
                    nombre = pelicula['titleText']['text']
                    año_estreno = pelicula['releaseYear']['year']
                    try:    
                        # Manejo del mes de estreno si está presente
                        if pelicula['releaseDate']['month'] is None:
                            mes_estreno = "-"
                        else:
                            mes_estreno = pelicula['releaseDate']['month']
                    except:
                        # Manejo de excepciones si no se proporciona el mes de estreno
                        mes_estreno = "-"
                    id_pelicula = pelicula['titleType']['id']
                    # Crear una tupla con la información de la película y agregarla a la lista
                    info_pelicula = (tipo_pelicula, nombre, año_estreno, mes_estreno, id_pelicula)
                    lista_peliculas.append(info_pelicula)
    # Solicitamos con return que nos 'devuelva' la lista de películas obtenidas
    return lista_peliculas

# Función para consultar y organizar los datos de películas en un DataFrame de pandas
def consultar_datos(genero, tipo, fecha_inicio, fecha_fin):
    # Obtenemos la lista de películas utilizando la función obtener_datos
    lista_peliculas = obtener_datos(genero, tipo, fecha_inicio, fecha_fin)
    # Creamos un DataFrame de pandas con los datos de las películas
    df = pd.DataFrame(lista_peliculas)
    # Devolvemos el DataFrame creado
    return df
```
